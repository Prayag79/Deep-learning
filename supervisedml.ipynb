```python
#Testing set
def split_train_test(data , test_ratio) :
  shuffeled_indices = np.random.permutation(len(data))
  test_set_size = int(len(data) * test_ratio)
  train_indices = shuffeled_indices[:-test_set_size]
  test_indices = shuffeled_indices[-test_set_size:]
  return data.iloc[train_indices] , data.iloc[test_indices]

```

```python
import hashlib

def test_set_check(identifier, test_ratio, hash=hashlib.md5):
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio


def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
  ids = data[id_column]
  in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
  return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
# we are using the id to uniqly identify , but if our data has some coloum which stays unique even for addition 0f new data , then we could instead use that but our data has no such attribute.
housing_with_id = df.reset_index()
train_set , test_set = split_train_test_by_id(housing_with_id , 0.2 , "index")
```
